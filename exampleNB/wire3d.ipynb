
# 3D wireframe plot


A very basic demonstration of a wireframe plot.



```python
data=[1,2,3,4,5,6,7,8,34,5,65,6]
re_shape = np.reshape(data, (3,4))
values=data[1:5]
print(values,re_shape)
```

```python
x=np.linspace(0,50,50)
y=np.linspace(0,50,50)
X, Y = np.meshgrid(x, y)
values = np.sqrt(X**2 + Y**2)
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(x, y, values)
plt.show()
```
